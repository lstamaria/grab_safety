# Executive Summary

This notebook attempts to predict if the driver is driving dangerously using telematic data. The workflow of this notebook follows the Cross-Industry Standard Process for Data Mining (CRISP-DM) framework to anchor the analytics solution to business value. The dataset is accessible from Grab website. It originally contains more than 16 million trips. After analysis, we can reduce the number to 20,000 trips only. ## Determine PCC

```python
from collections import Counter

fig, ax = plt.subplots(1, 1, figsize=(3,3), dpi=150)

state_counts = Counter(df['label'])
df_state = pd.DataFrame.from_dict(state_counts, orient='index')

tot = df_state[0].sum()
num = (df_state[0]/df_state[0].sum())**2
pcc = 1.25*num.sum()

ax = df_state.plot(kind='bar', title='Target\n1.25*PCC={0:0.4f}'.format(pcc), legend=None, 
              rot=0, ax=ax)

ax.set_xlabel('Dangerous Driving')
ax.set_ylabel('Number of Occurence');

print('Total: {:,}'.format(tot))
print('Population per class: {}\n'.format(df_state))
print('PCC={0:0.4f}'.format(num.sum()))
print('1.25*PCC={0:0.4f}'.format(pcc))
fig.tight_layout();
```

The Proportional Chance Criterion (PCC) indicates the score that is by random chance. Multiplying it to 1.25 increases that threshold that the prediction is not by random chance. If a predictive model is above $1.25 \times PCC$, it means that its prediction is performing above random chance only. We do this in order to minimize uncertainty inn prediction.

The $1.25 \times PCC$ is 0.7817. This means that we need to achieve an accuracy of at least 78.17\% to make sure that the predictive model performs above random chance. 

Getting a score at least $1.25 \times PCC$  will be serve as the basis of evaluating the predictive model. The succeeding sections provide detailed sequence from sampling to evaluation.

# Data Preparation

## Create derived features

Derived features are features aggregated from raw features. For example, we can combine `acceleration_x`, `acceleration_y`, and `acceleration_z` into `acceleration` to represent them.

We perform this task in order to reduce the number of features. This is important so that we can reduce the dimensionality of our dataset. Dimensionality reduction could help in improving the performance of our model because it can focus on relatively few features to make a prediction. The formula below provides a way to combine the features as one based on the euclidian distance of 3 orthogonal vectors.

\begin{equation}\label{accel}
a = \sqrt{a_x^2 + a_y^2 + a_z^2}
\end{equation}

Equation \ref{accel} shows how 3 vectors are combined as one using Euclidian formula. We apply this formula to `accelerometer` and `gyro` as derived features.

```python
df2 = df.copy()
```

```python
df2['acceleration'] = np.sqrt(df2['acceleration_x']**2 + df2['acceleration_y']**2 + df2['acceleration_z']**2)
```

```python
df2 = df2.drop(['acceleration_x', 'acceleration_y', 'acceleration_z'], axis=1)
```

```python
df2['gyro'] = np.sqrt(df2['gyro_x']**2 + df2['gyro_y']**2 + df2['gyro_z']**2)
```

```python
df2 = df2.drop(['gyro_x', 'gyro_y', 'gyro_z'], axis=1)
```

# Modeling

This section covers the approach to build a model using classification algorithm. The basis of using `GradientBoostingClassifier` is anchored on the fact that it performs well in most classification problems and provides explainability of top features using feature importance.

In addition, we perform the following to optimize the model:

1. Scaling such as `Normalizer` from `sklearn` to normalize the training set
1. Synthetic Minority Oversample Technique or `SMOTE` to address the imbalance of the data
1. `GridsearchCV`to find the hyperparameters of the best model
1. k-fold cross-validation incorporated in `GridSearchCV` to perform to improve the performance of the model

```python
# Define optimal threshold ROC curve
from sklearn.metrics import auc

def plot_threshold(version, y_test, y_pred, y_proba):
    
    fig, ax = plt.subplots(1, 1, figsize=(3,3), dpi=150)
    
    # Calculate ROC AUC score
    roc_auc = roc_auc_score(y_test, y_proba)
    # print('AUC: %.3f' % auc)
    
    # Calculate roc curve
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    
    # Plot no skill
    ax.plot([0, 1], [0, 1], linestyle='--', color='gray', alpha=0.3)
    
    # Plot ROC curve for the model
    ax.plot(fpr, tpr, marker=None, color='gray', alpha=0.3)
    
    # Find optimal threshold
    roc = pd.DataFrame({'fpr':fpr, 'tpr':tpr, 'thresholds':thresholds})
    roc['distance'] = np.sqrt((1-roc['tpr'])**2 + ((roc['fpr']))**2)
    optimal = roc[roc['distance']==roc['distance'].min()]
    ox = float(optimal['fpr'].values)
    oy = float(optimal['tpr'].values)
    
    ax.plot([0, ox], [1, oy], linestyle=':', color='#ff7f0e')
    
    ax.plot([0, 0], [oy, 1], linestyle=':', color='#1f77b4')
    ax.plot([0, ox], [oy, oy], linestyle=':', color='#1f77b4')
    ax.plot([0, ox], [1, 1], linestyle=':', color='#1f77b4')
    ax.plot([ox, ox], [oy, 1], linestyle=':', color='#1f77b4')
    ax.plot(ox, oy, linestyle='--', color='#1f77b4', marker='.')
    
    # show the plot
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_ylim([0.0, 1.05])
    ax.set_xlim([-0.05, 1.0])
    ax.set_title('ROC-AUC={0:0.3f}'.format(roc_auc))
    print('ROC-AUC={0:0.3f}'.format(roc_auc))
    
    fig.tight_layout()
    
    return optimal
```

```python
# Define ROC curve
from sklearn.metrics import auc

def plot_roccurve(version, y_test, y_pred, y_proba):
    
    fig, ax = plt.subplots(1, 1, figsize=(3,3), dpi=150)
    
    # Calculate ROC AUC score
    roc_auc = roc_auc_score(y_test, y_proba)
    # print('AUC: %.3f' % auc)
    
    # Calculate roc curve
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    
    # Plot no skill
    ax.plot([0, 1], [0, 1], linestyle='--')
    
    # Plot ROC curve for the model
    ax.plot(fpr, tpr, marker='.')
    
    # show the plot
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_ylim([0.0, 1.05])
    ax.set_xlim([-0.05, 1.0])
    ax.set_title('ROC-AUC={0:0.3f}'.format(roc_auc))
    print('ROC-AUC={0:0.3f}'.format(roc_auc))
    fig.tight_layout();
```

```python
# Define PR curve plot
def plot_prcurve(version, y_test, y_pred, y_proba):
    
    fig, ax = plt.subplots(1, 1, figsize=(3,3), dpi=150)
    
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    auc_score = auc(recall, precision)
    average_precision = average_precision_score(y_test, y_proba)
    
    print('f1=%.3f  auc=%.3f  average_precision=%.3f' % (f1, 
                                                         auc_score,
                                                         average_precision))
    step_kwargs = ({'step': 'post'}
                   if 'step' in signature(ax.fill_between).parameters
                   else {})
    ax.step(recall, precision, alpha=1, where='post', color='#ff7f0e', marker='.')
    ax.fill_between(recall, precision, alpha=0.7, **step_kwargs, color='#ff7f0e')
    
    ax.axhline(y=df_state[0][1] / df_state[0].sum(), linestyle='--', color='#1f77b4')

    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    ax.set_ylim([0.0, 1.05])
    ax.set_xlim([-0.05, 1.0])
    ax.set_title('AP=%.3f, AUC=%.3f'%(average_precision, auc_score))
    print('AP=%.3f, AUC=%.3f'%(average_precision, auc_score))
    fig.tight_layout();
```

```python
# Define SMOTE + GBM GridSearch 
def gbm_multiple(X, y, scaler=None, smote='auto', cv=None, refit='auc'):
    # Set testing and training sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    
    # Normalize
    if scaler=='norm':
        X_train = Normalizer().fit_transform(X_train)
    
    # Apply SMOTE
    sm = SMOTE(random_state=42, sampling_strategy=smote)
    X_train, y_train = sm.fit_sample(X_train, y_train)
    
    # Set param_grid
    param_grid = {'learning_rate': [0.1, 0.05, 0.01, 0.005],
                   'max_depth': [5, 6, 7, 8],
                   'min_samples_leaf': [4, 5, 6, 7],
                   'max_features': [.5, .3, .2]}
              
    scoring = {'auc': 'roc_auc',
               'accuracy': make_scorer(accuracy_score),
               'precision': make_scorer(precision_score),
               'recall': make_scorer(recall_score),
               'f1': make_scorer(f1_score)
              }

    gs = GridSearchCV(estimator=GradientBoostingClassifier(random_state=42),
                      param_grid=param_grid, cv=cv, scoring=scoring, 
                      refit=refit, return_train_score=True, n_jobs=-1)
    
    gs.fit(X_train, y_train)
    
    estimator = gs.best_estimator_
    score = gs.best_score_
    params = gs.best_params_
    index = gs.best_index_
    
    results = gs.cv_results_
    
    y_pred = gs.best_estimator_.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    confmatrix = confusion_matrix(y_test, y_pred)
    classreport = classification_report(y_test, y_pred)
    
    y_proba = gs.best_estimator_.predict_proba(X_test)
    y_proba = y_proba[:, 1]
    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    auc_score = auc(recall, precision)
    aveprecision = average_precision_score(y_test, y_proba)
    
    return {'estimator':estimator,'score':score, 'params':params, 
            'index':index, 'scoring':scoring, 'results':results,  
            'X_train':X_train,'y_train':y_train,
            'X_test':X_test,'y_test':y_test, 'y_pred':y_pred, 'y_proba':y_proba,
            'confmatrix':confmatrix, 'classreport':classreport, 
            'smote':smote, 'cv':cv, 'refit':refit,
            'accuracy':accuracy, 'auc':auc_score, 'f1':f1, 'aveprecision':aveprecision}
```

## Summary of Results

Initially, the model with highest accuracy against PCC is determined as the best model. However, we look into other measures to investigate model performance further. We perform this to anchor the value the solution to the business value at hand: the safety of drivers, including passengers.

The measures that we can look into are the following:

1. ROC-AUC score indicates the area under curve (AUC) of the Receiver-operating characteristic (ROC) curve. It is derived getting the integral of the area of the ROC curve. It is based on two metrics: true positive rate (TPR) and false positive rate (FPR). A higher value means better performance. Below are the formulas.

\begin{equation}\label{tpr}
\frac{TP}{(TP+FP)}
\end{equation}

\begin{equation}\label{fpr}
1 - \frac{TN}{(TN+FP)}
\end{equation}

Equation \ref{tpr} refers to true positive rate (TPR), how well the model predicts true positives in the midst of false positives (FP). Equation \ref{fpr} refers to false positive rate (TPR), a derived equation which indicates how well the model predicts false positives in the midst of true negatives (TN).

We can focus our attention on how the model performs in the face of false negatives (the trip we thought did not involve dangerous driving but it did). We would like a lower occurrence of false negatives. This is important because we need the model to become *sensitive* in predicting dangerous driving and minimize the *miss rate*, the false negative rate.

Given that, we employ the following techniques:

1. Receiver-operating characteristic (ROC) curve that looks into true positive rate (TPR) and false positive rate (FPR). TPR represents how good the model is in predicting true positives in the mix of false positives.
1. Precision-recall (PR) curve that looks into precision and recall. Recall represents how good the model is in predicting true positives in the mix of false negatives.
